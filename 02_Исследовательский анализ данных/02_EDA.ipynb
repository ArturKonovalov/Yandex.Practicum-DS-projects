# Исследование объявлений о продаже квартир

В нашем распоряжении данные сервиса Яндекс.Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктов за несколько лет. 
Нужно научиться определять рыночную стоимость объектов недвижимости. Наша задача — установить параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые — получены автоматически на основе картографических данных.

**Цель исследования** — выполнить предобработку данных и изучить их, чтобы найти интересные особенности и зависимости, которые существуют на рынке недвижимости. 

**Также рассмотрим несколько вопросов:**
1.	Как быстро продавались квартиры?
2.	Какие факторы больше всего влияют на общую (полную) стоимость объекта?.
3.	Какова средняя цена одного квадратного метра в 10 населённых пунктах с наибольшим числом объявлений?
4.	Какова средняя цена каждого километра и как стоимость объектов зависит от расстояния до центра города?

**Ход исследования**

```python
print(
    'living_area < 18:',
    len(data[data['living_area'] < 18]) # подставляем значения 1-го  квартиля
) 
print(
    'living_area > 42:',
    len(data[data['living_area'] > 42]) # подставляем значения 3-го квартиля
) 
data_area = data[['total_area', 'living_area', 'kitchen_area', 'studio', 'is_apartment']]

print(
    'Площадь кухни > жилой площади:',
    len(data_area[data_area['living_area'] < data['kitchen_area']])
)
data_area[data_area['living_area'] < data['kitchen_area']].head()
```

*В крайней таблице наблюдаем, что площадь кухни больше жилой площади, поменяем значения местами:*

```python
#циклом меняем значения living_area и kitchen_area местами, если kitchen_area > living_area
pd.set_option('chained_assignment', None)
for index in range(len(data)): 
    if data['living_area'][index] < data['kitchen_area'][index]:
        data['living_area'][index], data['kitchen_area'][index] =\
        data['kitchen_area'][index], data['living_area'][index]

data_area = data[['total_area', 'living_area', 'kitchen_area', 'studio', 'is_apartment']]

print(
    'Площадь кухни > жилой площади:',
    len(data_area[data_area['living_area'] < data['kitchen_area']])
)

data[['total_area', 'living_area', 'kitchen_area', 'rooms', 'last_price']].corr()
```

```python
data['total_area'].describe()
data_area[data_area['total_area'] < 38].count()
data['total_area_id'] = pd.qcut(data['total_area'], q=30) 
data['total_area_id'].unique()
```

*Заполняем пропуски в `living_area` и `kitchen_area` медианами, группированных по категориям данных:*

```python
#Так как квартира может быть студией, то площадь кухонь у них заменим на 0
data.loc[data['studio'] == True, 'kitchen_area'] = 0 

data['living_area'] = data.groupby('total_area_id')['living_area'].apply(
    lambda s: s.fillna(s.median())
)
print('Количество Nan в living_area', data['living_area'].isna().sum())
data['living_area'].hist(
    bins=15,
    range=(0,60),
    alpha=0.4
)

data['kitchen_area'] = data.groupby('total_area_id')['kitchen_area'].apply(
    lambda s: s.fillna(s.median())
)
print('Количество Nan в kitchen_area', data['kitchen_area'].isna().sum())
data['kitchen_area'].hist(
    bins=15,
    range=(0,60),
    alpha=0.4
)

plt.xlabel('Площадь', fontsize=14)
plt.ylabel('Значения', fontsize=14)
plt.show()
```

*Удалим максимально крайние значения для всех столбцов, задав условия нахождения значений в диапазоне квантиля от `0.005` до `0,0995`:*

```python
filtered_columns = [
    'last_price','total_area', 'floors_total',
    'living_area', 'kitchen_area', 'days_exposition',
    'airports_nearest', 'city_centers_nearest', 
    'parks_nearest', 'ponds_nearest'
]

for column in filtered_columns:
    data = data[
        (data[column].quantile(0.005) <= data[column]) & 
        (data[column].quantile(0.995) >= data[column])
    ] 
```
