# **Исследование факторов, влиящих на успешность игр**

**Описание проекта:**
Интернет-магазин «Стримчик» продаёт по всему миру компьютерные игры. Заказчик хочет выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

**Описание данных:** В нашем распоряжении исторические данные с 1980 по 2016 года о продажах игр, оценки пользователей и экспертов, жанры и платформы, взятые из открытых источников. 

**Цель исследования** — выполнить предобработку данных и изучить их, чтобы найти интересные особенности и зависимости, выявить определяющие успешность игры закономерности.

**Также рассмотрим несколько вопросов:**
1.	Сколько игр выпускалось в разные годы?

### Заполнение пропусков и преобразование данных в нужные типы.

#### Заполнение пропусков в `name` и `genre`:

Пропуски могли возникнуть случайно.

*Т.к пропуски в этих столбцах составляют всего лишь по 2 единицы в каждом, то отсутствующие значения просто удалим*

```python
data = data.dropna(subset=['name', 'genre'])
print(data['name'].isna().sum())
print(data['genre'].isna().sum())
```

___
#### Заполнение пропусков и преобразование данных в `year_of_release`:

*Возможно пропуски в данном столбце вызваны тем, что игра не имела широкого распространения, так сказать, была 'noname', поэтому информации о дате выпуска в принципе не существует*

Т.к кол-во пропусков всего 2% , то данные строки удалим.

```python
data['year_of_release'].describe()
```

```python
data['year_of_release'].isna().sum()
```

```python
data = data.dropna(subset = ['year_of_release'])
data['year_of_release'].isna().sum()
```

*Преобразуем данный столбец в тип `int`, так как года исчисляются целочисленно:*

```python
data['year_of_release'] = data['year_of_release'].astype('Int64')
data['year_of_release'].head()
```

___
####  Преобразование данных в  `user_score`:

Т.к заполнение пропусков может лишь замусорить данные, оставим значения в нетронутом виде.

```python
data['user_score'].unique()
```

Неопределeнный тип значений `tbd`, (to be determined) — изначально не подлежат никакой категории, их необходимо заполнить. Так как природа таких неясностей нам неизвестна, то будем считать, что tbd - пропущенное значение и заменим их на `NaN`.

```python
data['user_score'].isna().sum()
```

```python
for d in data['user_score']:
    data.loc[(data['user_score'] == 'tbd'), 'user_score'] = np.nan
    
data['user_score'].isna().sum()
```

*Оценки пользователей - это численные знаечния, поэтому `object` меняем на `float`*

```python
data['user_score'] = data['user_score'].astype(float)
data['user_score'].describe()
```

___
####  Заполнение пропусков и преобразование данных в  `critic_score`:

*Пропуски могут быть связаны с тем, что игра не стала популярной. Заполнять ничем не будем, оставим в первозданном виде.

Возможно все критики в это время рецензировали `Cyber Pank` и так до них и не добрались*

*Оценки критиков не имеют сотых долей исчесления, поэтому переводим столбец к типу `int`*

```python
data['critic_score'] = data['critic_score'].astype('Int64')
data['critic_score'].head()
```

___
####  Заполнение пропусков  `rating`:

Т.к. локальные разработчики игр и небольшие игровые студии могут быть не включены в рейтинг *ESRB*, то заполним пропуски "заглушкой", чтобы игры без рейтинга не выплаи из исследования.

```python
data['rating'].isna().sum()
```

```python
data['rating'] = data['rating'].fillna('unknown')
data['rating'].isna().sum()
```

Посчитаем суммарные продажи во всех регионах и запишем их в отдельный столбец.
